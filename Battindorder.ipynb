# Converting Runs to Wins

Runs per win (RPW) is the value that allows us to convert runs to wins. 

It is based on the run environment and is usually between 9 and 10.

RPW = 9*(MLB Runs Scored / MLB Innings Pitched)*1.5 + 3

# Class Player

```python
class Player:
    def __init__(self, pos, name, g, pa, ab, r, h, dbl, tpl, 
                 hr, rbi, sb, cs, bb, so, hbp, sh, sf, ibb, 
                 xbt, gb_fb, prdout, gidp, sbo):
        self.pos = pos
        self.name = name
        self.g = g
        self.pa = pa
        self.ab = ab
        self.r = r
        self.h = h
        self.dbl = dbl
        self.tpl = tpl
        self.hr = hr
        self.rbi = rbi
        self.sb = sb
        self.cs = cs
        self.bb = bb
        self.so = so
        self.hbp = hbp
        self.sh = sh
        self.sf = sf
        self.ibb = ibb
        self.xbt = xbt
        self.gb_fb = gb_fb
        self.prdout = prdout
        self.gidp = gidp
        self.sbo = sbo
        
        # Create derived stats
        self.obp = (self.h+self.bb+self.hbp+self.ibb)/(self.pa)
        self.slg = (self.h+self.dbl+(2*self.tpl) + (3*self.hr))/(self.ab)
        self.ops = self.obp+self.slg
        if (self.sb + self.cs > 0) & (self.sbo > 0):
            self.sbt = (self.sb+self.cs)/(self.sbo)
            self.sbs = (self.sb)/(self.sb+self.cs)
        else:
            self.sbt = 0.0
            self.sbs = 0.0
    
    def __repr__(self):
        return('{}'.format(self.name)) 
    
    def __str__(self):
        return('{}'.format(self.name)) 
        
    # Run outputs the result of the movement of a baserunner
    # Inputs:
    #         start: the base the runner started on
    #        action: the reason the runner was moving, 
    #                the output of running PlateAppearance
    # Output:
    #        1.5: out at second base
    #        2.0: safe at second base
    #        2.5: out at third base
    #        3.0: safe at third base
    #        3.5: out at home
    #        4.0: safe at home, run scored
    def Run(self, start, action):
        returnValue = start
        extra = np.random.random()
        out = np.random.random()
        
        # 1.2 and 1.4 involves dead ball, runner advances
        if action % 1 > 0:
            return(returnValue + 1)
        # Runner scores on a triple or home run, 
        # or a double when starting at 2nd
        if (action > 2 or (action == 2 and start >= 2)):
            return(4.0)
        
        returnValue = action + start
        # See if the runner took an extra base
        if extra <= self.xbt:
            returnValue += 1.0
        
        # 4 is the highest value allowed
        if returnValue > 4.0:
            returnValue = 4.0
            
        # Sometimes a runner is thrown out
        if out <= 0.05:
            returnValue -= 0.5
        
        return(returnValue)
        
        
    
    # PlateAppearance outputs the result of a hitter's plate appearance
    # Inputs:
    #         none
    # Output: 
    #         0.0: out
    #         1.0: single
    #         1.2: walk
    #         1.4: hit by pitch
    #         2.0: double
    #         3.0: triple
    #         4.0: home run
    
    def PlateAppearance(self):
        returnValue = 0.0
        ba = self.h/self.pa
        obp = (self.h + self.hbp + self.bb)/self.pa
        result = np.random.random()
        if result <= ba:
            returnValue = 1.0
            homerun = 1-(self.hr/self.h)
            triple = homerun - (self.tpl/self.h)
            double = triple - (self.dbl/self.h)
            x = np.random.random()
            if x > homerun:
                returnValue += 3
            elif x > triple:
                returnValue += 2
            elif x > double:
                returnValue += 1
        elif result <= obp:
            walk = self.bb/(self.bb+self.hbp)
            x = np.random.random()
            if x < walk:
                returnValue = 1.2 # Walk
            else:
                returnValue = 1.4 # Hit by pitch
        else:
            if result > (1 - self.so/self.pa):
                returnValue = 0.0
            elif result > (1 - self.gidp-self.so/self.pa):
                x = np.random.random()
                if (x > self.prdout):
                    returnValue = -2.0 # double play if possible and no productOut made
                else:
                    returnValue = -1.8 # no runner on first base and productOut made
            else:
                x = np.random.random()
                y = np.random.random()
                if x < (1 - 1/(1+self.gb_fb)):
                    if (y < self.prdout):
                        returnValue = -1.2 # productOut mady by groundball if runner on base
                    else:
                        returnValue = 1.25
                else:
                    if (y < self.prdout):
                        returnValue = -1.4 # productOut maded by flyball if runner on base with second or third
                    else:
                        returnValue = 0.0
            
        return(returnValue)
    
    def Stealing(self):
        returnValue = 0.0
        if np.random.random() < self.sbt:
            if np.random.random() < self.sbs:
                returnValue = 1.0
            else:
                returnValue = -1.0
        else:
            returnValue = 0.0
        return(returnValue)
                
    # ShowStats returns a dataframe with player's name and statistics
    # Stats: G, PA, AB, R, H, Dbl, Tpl, HR, RBI, SB, CS, BB, SO, HBP
    def ShowStats(self):
        d = {name: pd.Series([self.g, self.pa, self.ab, self.r, self.h,
                              self.dbl, self.tpl, self.hr, self.rbi, 
                              self.sb, self.cs, self.bb, self.so, self.hbp],
                            index=['G', 'PA', 'AB', 'R', 'H', '2B', '3B', 
                                   'HR', 'RBI', 'SB', 'CS', 'BB', 'SO', 
                                   'HBP'])}
        return(pd.DataFrame(d))    
```

# Game Simulator

```python
def simulate(hitters):
    inning = 1
    runs = 0
    currentHitter = 0
   
    while inning <= 9:
        # Start the inning with no outs and the  bases clear
        outs = 0
        # Runners on first, second, and third
        runner = ['empty','empty','empty'] 
        
        while outs < 3:
            pa = hitters[currentHitter].PlateAppearance()
            # League fielding pct is .985 according to Baseball Reference
            error = 1 if (np.random.random() > .985) else 0 
            
            #Stealing bases
            if (runner[0]!='empty') & (runner[1]=='empty'):
                st = runner[0].Stealing()
                if st == 1.0:
                    runner[1] = runner[0]
                    runner[0] = 'empty'
                elif st == -1.0:
                    runner[0] = 'empty'
                    outs += 1
                else:
                    runner[0] = runner[0]
            
            if pa > 0:
                # Hitter got on base, move runners 
                # and put the hitter on base
                extraBase = pa+1 if (np.random.random() < .7) else 0
                # Move runner from third, then from second, then from first
                for x in [2,1,0]: 
                    if runner[x] != 'empty':
                        result = runner[x].Run(x,pa) + extraBase + error
                        if (result % 1 > 0) & (error == 0): # Runner is out
                            outs += 1
                        elif result >= 4.0: # Runner scored
                            runs += 1
                        else:
                            # Base is occupied, runner cannot advance
                            if runner[int(result)-1] != 'empty': 
                                result -= 1
                                runner[int(result)-1] = runner[x]
                            else:
                                runner[int(result)-1] = runner[x]
                        runner[x] = 'empty'
                if pa >= 4:
                    runs += 1
                else:
                    runner[int(pa)-1] = hitters[currentHitter]
            else:
                if error > 0: 
                    if runner[2] != 'empty':
                        runs +=1
                        runner[2] = 'empty'
                    if runner[1] != 'empty':
                        runner[2] = runner[1]
                        runner[1] = 'empty'
                    if runner[0] != 'empty':
                        runner[1] = runner[0]
                    runner[0] = hitters[currentHitter]
                else:
                    if outs == 0:
                        if (runner[0] != 'empty') & (runner[1] == 'empty') & (runner[2] == 'empty'):
                            if (pa == -2.0) | (pa == -1.8):
                                runner[0] = 'empty'
                                outs += 2
                            elif pa == -1.2:
                                runner[1] = runner[0]
                                runner[0] = 'empty'
                                outs += 1
                            elif pa == -1.25:
                                runner[0] = hitters[currentHitter]
                                outs += 1
                            else:
                                outs +=1
                        elif (runner[0] == 'empty') & (runner[1] != 'empty') & (runner[2] == 'empty'):
                            if (pa == -1.8) | (pa == -1.2) | (pa == -1.4):
                                runner[2] = runner[1]
                                runner[1] = 'empty'
                                outs += 1
                            else:
                                outs += 1
                        elif (runner[0] == 'empty') & (runner[1] == 'empty') & (runner[2] != 'empty'):
                            if (pa == -1.8) | (pa == -1.2) | (pa == -1.4):
                                runner[2] = 'empty'
                                runs += 1
                                outs += 1
                            else:
                                outs += 1
                        elif (runner[0] != 'empty') & (runner[1] != 'empty') & (runner[2] == 'empty'):
                            if (pa == -2.0) | (pa == -1.8):
                                runner[2] = runner[1]
                                runner[1] = 'empty'
                                runner[0] = 'empty'
                                outs += 2
                            elif pa == -1.2:
                                if np.random.random() > 0.5:
                                    runner[2] = runner[1]
                                    runner[1] = runner[0]
                                    runner[0] = 'empty'
                                    outs += 1
                                else:
                                    runner[2] = runner[1]
                                    runner[1] = 'empty'
                                    runner[0] = hitters[currentHitter]
                                    outs += 1
                            elif pa == -1.25:
                                runner[1] = runner[0]
                                runner[0] = hitters[currentHitter]
                                outs += 1
                            elif pa == -1.4:
                                runner[2] = runner[1]
                                runner[1] = 'empty'
                                outs += 1
                            else:
                                outs += 1
                        elif (runner[0] != 'empty') & (runner[1] == 'empty') & (runner[2] != 'empty'):
                            if (pa == -2.0) | (pa == -1.8):
                                runner[2] = 'empty'
                                runner[0] = 'empty'
                                runs += 1
                                outs += 2
                            elif pa == -1.2:
                                runs += 1
                                outs += 1
                                if np.random.random() > 0.5:
                                    runner[2] = 'empty'
                                    runner[1] = runner[0]
                                    runner[0] = 'empty'
                                else:
                                    runner[2] = 'empty'
                                    runner[1] = 'empty'
                                    runner[0] = hitters[currentHitter]
                            elif pa == -1.4:
                                runner[2] = 'empty'
                                runs += 1
                                outs += 1
                            else:
                                outs += 1
                        elif (runner[0] == 'empty') & (runner[1] != 'empty') & (runner[2] != 'empty'):
                            if (pa == -1.8) | (pa == -1.2) | (pa == -1.4):
                                runs += 1
                                outs += 1
                                if np.random.random() >0.5:
                                    runner[2] = runner[1]
                                    runner[1] = 'empty'
                                else:
                                    runner[2] = 'empty'
                            else:
                                outs +=1
                        elif (runner[0] != 'empty') & (runner[1] != 'empty') & (runner[2] != 'empty'):
                            if (pa == -2.0) | (pa == -1.8):
                                runner[2] = runner[1]
                                runner[1] = 'empty'
                                runner[0] = 'empty'
                                runs += 1
                                outs += 2
                            elif pa == -1.2:
                                runner[2] = runner[1]
                                runner[1] = runner[0]
                                runner[0] = 'empty'
                                runs += 1
                                outs += 1
                            elif pa == -1.4:
                                runner[2] = runner[1]
                                runner[1] = 'empty'
                                runs += 1
                                outs += 1
                            elif pa == -1.25:
                                runner[2] = runner[1]
                                runner[1] = runner[0]
                                runner[0] = hitters[currentHitter]
                                outs += 1
                            else: 
                                outs += 1
                        else:
                            outs += 1
                    elif outs == 1:
                        if (runner[0] != 'empty') & (runner[1] == 'empty') & (runner[2] == 'empty'):
                            if (pa == -2.0) | (pa == -1.8):
                                outs += 2
                            elif pa == -1.2:
                                runner[1] = runner[0]
                                runner[0] = 'empty'
                                outs +=1
                            elif pa == -1.25:
                                runner[0] = hitters[currentHitter]
                                outs += 1
                            else:
                                outs +=1
                        elif (runner[0] == 'empty') & (runner[1] != 'empty') & (runner[2] == 'empty'):
                            if (pa == -1.8) | (pa == -1.2) | (pa == -1.4):
                                runner[2] = runner[1]
                                runner[1] = 'empty'
                                outs += 1
                            else:
                                outs += 1
                        elif (runner[0] == 'empty') & (runner[1] == 'empty') & (runner[2] != 'empty'):
                            if (pa == -1.8) | (pa == -1.2) | (pa == -1.4):
                                runs += 1
                                runner[2] = 'empty'
                                outs += 1
                            else:
                                outs += 1
                        elif (runner[0] != 'empty') & (runner[1] != 'empty') & (runner[2] == 'empty'):
                            if (pa == -2.0) | (pa == -1.8):
                                outs += 2
                            elif pa == -1.2:
                                if np.random.random() > 0.5:
                                    runner[2] = runner[1]
                                    runner[1] = runner[0]
                                    runner[0] = 'empty'
                                    outs += 1
                                else:
                                    runner[2] = runner[1]
                                    runner[1] = 'empty'
                                    outs += 1
                            elif pa == -1.25:
                                runner[1] = runner[0]
                                runner[0] = hitters[currentHitter]
                                outs += 1
                            elif pa == -1.4:
                                runner[2] = runner[1]
                                runner[1] = 'empty'
                                outs += 1
                            else:
                                outs += 1
                        elif (runner[0] != 'empty') & (runner[1] == 'empty') & (runner[2] != 'empty'):
                            if (pa == -2.0) | (pa == -1.8):
                                outs += 2
                            elif pa == -1.2:
                                runs += 1
                                outs += 1
                                if np.random.random() > 0.5:
                                    runner[2] = 'empty'
                                    runner[1] = runner[0]
                                    runner[0] = 'empty'
                                else:
                                    runner[2] = 'empty'
                                    runner[1] = 'empty'
                                    runner[0] = hitters[currentHitter]
                            elif pa == -1.4:
                                runner[2] = 'empty'
                                runs += 1
                                outs += 1
                            else:
                                outs += 1
                        elif (runner[0] == 'empty') & (runner[1] != 'empty') & (runner[2] != 'empty'):
                            if (pa == -1.8) | (pa == -1.2) | (pa == -1.4):
                                runs += 1
                                outs += 1
                                if random.random() > 0.5:
                                    runner[2] = runner[1]
                                    runner[1] = 'empty'
                                else:
                                    runner[2] = 'empty'
                            else:
                                outs +=1
                        elif (runner[0] != 'empty') & (runner[1] != 'empty') & (runner[2] != 'empty'):
                            if (pa == -2.0) | (pa == -1.8):
                                outs += 2
                            elif pa == -1.2:
                                runner[2] = runner[1]
                                runner[1] = runner[0]
                                runner[0] = 'empty'
                                runs += 1
                                outs += 1
                            elif pa == -1.4:
                                runner[2] = runner[1]
                                runner[1] = 'empty'
                                runs += 1
                                outs += 1
                            elif pa == -1.25:
                                runner[2] = runner[1]
                                runner[1] = runner[0]
                                runner[0] = hitters[currentHitter]
                                outs += 1
                            else:
                                outs += 1
                        else:
                            outs += 1
                    else:
                        outs += 1
                                
            # Move on to the next hitter, 
            # going to the top of the lineup after the last
            currentHitter += 1
            if currentHitter > (len(hitters)-1):
                currentHitter = 0
                
        inning += 1 # Inning over, go to the next inning
    
    return(runs)
```

# Conventional Batting Order Construction

The first or leadoff hitter should be good at getting on base and preferably a good baserunner as well. He should be willing to watch many pitches so that his teammates get a better chance to see the opposing pitcher's stuff. It is usually a waste to put a power hitter in the leadoff spot, although there have been notable exceptions.

In the traditional approach, the second hitter should be a good bat handler. He needs to be able to take pitches to give the leadoff man a chance to steal. He should also be able to make a sacrifice bunt or hit and run play. Another theory is that any player with a high on-base percentage should occupy the spot, no matter his other skills, in order to create RBI opportunities for the team's best sluggers who normally hit in the next two spots.

The third hitter is supposed to be the best all-around hitter on the team. He should be able to hit for average and power, and ideally should be able to run the bases well.

The fourth or cleanup hitter is supposed to be the best power hitter. His job is to drive in the top three hitters when they get on base.

The fifth place hitter is usually another power hitter, but one who isn't quite as good as the cleanup hitter.

The sixth place hitter is something like a second leadoff hitter. If the team has a second player with leadoff-type skills, he'll often bat 6th.

The seventh place batter is normally a spot for a batter who lacks the skills that would put him higher in the order.

The eighth place hitter's role depends on the league. In leagues that use the designated hitter, he is often viewed similarly to the seventh place hitter. In leagues in which the pitcher is required to bat, teams will often pitch around the eighth-place hitter, so it's desirable to have a patient hitter there. It used to be traditional to bat the catcher eighth because the catcher was often changed along with the pitcher.

The ninth spot is traditionally reserved for the pitcher in non-DH leagues. In DH leagues, it is often seen as a second leadoff spot, so teams will pick a batter with leadoff-type skills. However, certain managers prefer to bat their pitcher eighth, in order to have a "second lead-off hitter" batting 9th, who will be on base more often when the line-up turns around.

```python
def setCWLineup(starters):
    returnValue = [None]*9
    ops = []
    obp = []
    slg = []
    krate = []
    wrate = []
    
    # Removes the batter from the possibilities
    # Removes the chance the same batter will be put into multiple spots
    def removeBatter(x):
        ops.pop(x)
        obp.pop(x)
        slg.pop(x)
        krate.pop(x)
        wrate.pop(x)
    
    # Create lists for relative stats
    for x in range(len(starters)):
        ops.append(starters[x].ops)
        obp.append(starters[x].obp)
        slg.append(starters[x].slg)
        krate.append(starters[x].so/starters[x].ab)
        wrate.append((starters[x].bb+starters[x].ibb)/starters[x].pa)
    
    lineupCopy = starters.copy()
    if NL == True:
        # The last hitter is the pitcher
        returnValue[8] = lineupCopy.pop(0)
        removeBatter(0)
        
    # The three hitter is the one with the highest ops
    hitter = ops.index(max(ops))
    returnValue[2] = lineupCopy.pop(hitter)
    removeBatter(hitter)
    
    # The leadoff hitter is the one with the highest obp
    hitter = obp.index(max(obp))
    returnValue[0] = lineupCopy.pop(hitter)
    removeBatter(hitter)
    
    # The two hitter is the one with the lowest strikeout rate
    hitter = krate.index(min(krate))
    returnValue[1] = lineupCopy.pop(hitter)
    removeBatter(hitter)
    
    # The cleanup hitter is the one with the highest slugging percentage
    hitter = slg.index(max(slg))
    returnValue[3] = lineupCopy.pop(hitter)
    removeBatter(hitter)
 
    # The five hitter is the one with the next highest slugging percentage
    hitter = slg.index(max(slg))
    returnValue[4] = lineupCopy.pop(hitter)
    removeBatter(hitter)    
    
    # The eight hitter is the one with the highest walk percentage
    hitter = wrate.index(max(wrate))
    returnValue[7] = lineupCopy.pop(hitter)
    removeBatter(hitter)
    
    # The number six hitter is the one with the highest obp
    hitter = obp.index(max(obp))
    returnValue[5] = lineupCopy.pop(hitter)
    removeBatter(hitter)
    
    if NL == True:
        # The seven hitter is the only one left for NL lineup
        returnValue[6] = lineupCopy[0]
    else:
        # The seven hitter is the one with the highest obp
        hitter = obp.index(max(obp))
        returnValue[6] = lineupCopy.pop(hitter)
        removeBatter(hitter)
        # The nine hitter is the only one left
        returnValue[8] = lineupCopy[0]
    
    return(returnValue)
```

# Simulated Annealing

```python
#  These are the variables that can be changed to optimize results
temp = 3500  # choose initial temperature around the beginning lineup runs
alpha = 0.95 # Used to decay temp
max_moves_no_improve = 1000 # How many moves that can be made without improvement before shutting down

def setSALineup(temp, alpha, max_moves_no_improve, starters):
    lineup_size = 9

    # Objective Function

    def lineup_runs(lineup,iterations):
        thisLineup = []
        for x in lineup:
            thisLineup.append(starters[x])
        return(simulateSeasons(thisLineup,iterations))


    # reverse a random tour segment
    def sub_tour_reversal(lineup):
        i, j = np.sort(np.random.choice(lineup_size, 2, replace=False))
        swapped = np.concatenate((lineup[0:i],
                                  lineup[j:-lineup_size + i - 1:-1],
                                  lineup[j + 1:lineup_size]))
        return [int(swapped[i]) for i in range(lineup_size)]

    # initialize with a random tour
    current_lineup = np.random.permutation(np.arange(lineup_size)).tolist()
    current_runs = lineup_runs(current_lineup,1)
    best_lineupSA = current_lineup
    best_runs = current_runs

    # Initialize the counter variables
    num_moves_no_improve = 0
    iteration = 1

    # These help with plotting the search convergence
    runsSA = [[iteration,current_runs]]
    runs_bestSA = [[iteration,best_runs]]

    startTime = time.time()
    while (num_moves_no_improve < max_moves_no_improve):

        num_moves_no_improve += 1
        new_lineup = sub_tour_reversal(current_lineup)
        new_runs = lineup_runs(new_lineup,3) 
        delta = current_runs - new_runs
        prob = np.exp(min(delta, 0) / temp)
        accept = new_runs > current_runs or np.random.uniform() < prob

        if accept:
            current_lineup = new_lineup
            current_runs = new_runs
            if current_runs > best_runs:
                current_runs = lineup_runs(current_lineup,10) 
                if current_runs > best_runs:
                    best_lineupSA = current_lineup
                    best_runs = current_runs
                    num_moves_no_improve = 0
        temp *= alpha
        iteration += 1
        runsSA.append([iteration,current_runs])
        runs_bestSA.append([iteration,best_runs])
        

    print('Completed on Iteration: {}'.format(iteration))
    print('Runs: {}, No Improve: {}'
          .format(best_runs,num_moves_no_improve))
    saTime = time.strftime("%H:%M:%S", time.gmtime(time.time()-startTime))
    print('Time Elapsed: {}'.format(saTime))
    
    return best_runs, best_lineupSA, saTime
    
    
```

```python
def setGALineup(pop_size, cx_prob, mut_prob, max_gen, max_no_improve, s):
    startTime = time.time()
    pop = toolbox.population(n=pop_size)
    logbook = tools.Logbook()
    hof = tools.HallOfFame(1)

    # Evaluate the entire population
    fitnesses = list(map(toolbox.evaluate, pop))
    for ind, fit in zip(pop, fitnesses):
        ind.fitness.values = fit

    hof.update(pop)
    best_val = hof[0].fitness.values
    num_no_improve = 0
    generation = 0

    while num_no_improve < max_no_improve and generation < max_gen:
         
        # Select the next generation individuals
        selected = toolbox.select(pop, len(pop))
        # Clone the selected individuals
        offspring = list(map(toolbox.clone, selected))

        # Apply crossover and mutation on the offspring
        for child1, child2 in zip(offspring[::2], offspring[1::2]):
            if random.random() < cx_prob:
                toolbox.mate(child1, child2)
                del child1.fitness.values
                del child2.fitness.values

        for mutant in offspring:
            if random.random() < mut_prob:
                toolbox.mutate(mutant)
                del mutant.fitness.values

        # Evaluate the individuals with an invalid fitness
        invalid_ind = [ind for ind in offspring if not ind.fitness.valid]
        fitnesses = map(toolbox.evaluate, invalid_ind)
        num_evals = 0
        for ind, fit in zip(invalid_ind, fitnesses):
            num_evals += 1
            ind.fitness.values = fit

        # The population is entirely replaced by the offspring
        pop[:] = offspring

        # track the best value and reset counter if there is a change
        hof.update(pop)
        curr_best_val = hof[0].fitness.values[0]
        num_no_improve += 1
        if curr_best_val != best_val:
            best_val = curr_best_val
            num_no_improve = 0

        # record stats
        record = stats.compile(pop)
        logbook.record(gen=generation, evals=num_evals, **record)

          
        # increment generation
        generation += 1
        
       
    print('Completed on Iteration: {}'.format(generation))
    print('Runs: {}, No Improve: {}'.format(best_val,num_no_improve))
    eTime = time.strftime("%H:%M:%S", time.gmtime(time.time()-startTime))
    print('Time Elapsed: {}'.format(eTime))
    best_x = list(hof[0])

    return best_val, best_x, eTime, logbook
    
# Objective Function
def lineup_runs(lineup,iterations):
    thisLineup = []
    for x in lineup:
        thisLineup.append(s[x])
    return(simulateSeasons(thisLineup,iterations))

def runs_tuple(individual, iterations):
    return (lineup_runs(individual, iterations),)

# create a maximization problem
creator.create("FitnessRuns", base.Fitness, weights=(1.0, ))

# tells us what kind of individuals we'll have - arrays of integers
creator.create("Individual",
               list,
               typecode='i',
               fitness=creator.FitnessRuns)

# configure toolbox
toolbox = base.Toolbox()

# define how we create an individual and a poopulation
def create_individual(ind_size):
    return random.sample(range(ind_size),ind_size) # choose a random permutation of 0 .... ind_size-1
toolbox.register("indices",create_individual,9)

toolbox.register("individual", tools.initIterate, creator.Individual,
                 toolbox.indices)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# set the fitness function
toolbox.register("evaluate", runs_tuple, iterations=50)

# Selection
toolbox.register("select", tools.selTournament, tournsize=3)
# Crossover
toolbox.register("mate", tools.cxPartialyMatched)
# Mutation
toolbox.register("mutate", tools.mutShuffleIndexes, indpb=0.1)

# configure statistics to collect
stats = tools.Statistics(key=lambda ind: ind.fitness.values)
stats.register("avg", np.mean)
stats.register("std", np.std)
stats.register("min", np.min)
stats.register("max", np.max)
```
